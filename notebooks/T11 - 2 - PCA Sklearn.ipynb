```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("../datasets/iris/iris.csv")
```

```python
X = df.iloc[:,:4].values
y = df.iloc[:,4].values
X_std = StandardScaler().fit_transform(X)
```

```python
acp = PCA(n_components=2)
Y = acp.fit_transform(X_std)
```

```python
import plotly.graph_objects as go

results = []
for name in ('setosa', 'versicolor', 'virginica'):   
    result = go.Scatter(x=Y[y==name,0], y = Y[y==name, 1],
                       mode = "markers", name=name,
                        marker=go.scatter.Marker(size = 15, opacity=0.8))
    results.append(result)

layout = go.Layout(showlegend=True, 
                   xaxis=go.layout.XAxis(title="CP 1"),
                   yaxis=go.layout.YAxis(title="CP 2")
                  )

fig = go.Figure(results, layout)
fig.show()
```

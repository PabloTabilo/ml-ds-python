# La simulación de Monte Carlo

Pbb de caer dentro de un circulo circunscrito a un rectangulo.
$$ P(caer dentro cir) = \frac{Casos favo cir}{todos los casos} = \frac{A_{cir}}{A_{rec}} = \frac{\pi r^2}{4 r^2} = \frac{\pi}{4} $$

```python
def pi_montecarlo(n_exp, n_lanzamientos):
    pi_vals = []
    for i in range(n_exp):
        casos_f = 0
        for i in range(n_lanzamientos):
            x = np.random.random()
            y = np.random.random()
            res = x**2 + y**2
            if(res <= 1):
                casos_f+=1
        pi_vals.append(4 * float(casos_f/n_lanzamientos+0.0))
    return(sum(pi_vals) / len(pi_vals))
```

```python
pi_montecarlo(n_exp=200, n_lanzamientos=10000)
```

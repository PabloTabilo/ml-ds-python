```python
np.random.seed(4711)
a = np.random.multivariate_normal([10,0], [[3,1],[1,4]], size=[100,])
b = np.random.multivariate_normal([0,20], [[3,1],[1,4]], size=[50,])
X = np.concatenate((a, b)) # apendizo dos distribuciones iguales con distinta media
print(X.shape)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, "ward")
```

## Cophenetic correlation
* The cophenetic correlation for a cluster tree is defined as the linear correlation coefficient between the cophenetic distances obtained from the tree, and the original distances (or dissimilarities) used to construct the tree. Thus, it is a measure of how faithfully the tree represents the dissimilarities among observations.

* is a measure of how faithfully a dendrogram preserves the pairwise distances between the original unmodeled data points.

$$ c = \frac{\sum_{i<j} (x(i,j) - \overline{x}) (t(i,j) - \overline{t})}{ \sqrt{\sum_{i<j} (x(i,j) - \overline{x})^2 \sum_{i<j}(t(i,j) - \overline{t})^2}} $$

* $ x(i, j) = | x_i - x_j | $, the euclidean distance between observartions.
* $ t(i,j) $ The dendrogrammatic distance between the model points $T_i$ and $T_j$. This distance is the height of the node at which these two points are first joined together.

```python
def dendogram_tune(*args, **kwargs):
    max_d = kwargs.pop("max_d", None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs["color_threshold"]=max_d
    annotate_above=kwargs.pop("annotate_above",0)
    ddata = dendrogram(*args, **kwargs)
    if not kwargs.get("no_plot", False):
        plt.title("Dendograma del clustering jerárquico")
        plt.xlabel("Indices de la dataset (o tamaño del cluster)")
        plt.ylabel("Distancia")
        for index, d, color in zip(ddata["icoord"], ddata["dcoord"], ddata["color_list"]):
            x = 0.5 * sum(index[1:3])
            y = d[1]
            if y>annotate_above:
                plt.plot(x,y,'o',c=color)
                plt.annotate("%.3g"%y,(x,y),xytext=(0,-5),
                            textcoords="offset points", va="top", ha="center")
    if max_d:
        plt.axhline(y=max_d, c="k")
    return ddata
```

```python
plt.figure(figsize=(10,8))
dendogram_tune(Z, truncate_mode="lastp", p=12, leaf_rotation=90.0,leaf_font_size=12.,show_contracted=True,
              annotate_above=10, #anotar distancias por encima de 10 de distancia eu
               max_d=20, #valor de corte
              )
plt.show()
```

# Métodos de corte de dendograma

## Métodos de inconsistencia

$$ inconsistency_i = \frac{h_i - \overline{h}}{s(h)} $$

Los parámetros son:
* $ h_i $ : Altura del cluster $i^{th}$
* $\overline{h}$ : Promedio muestral de las alturas de todos los clusters
* $ s(h) $ : Desviación estándar muestral de todos los clusters

```python
from scipy.cluster.hierarchy import inconsistent
```

```python
depth = 5
incons = inconsistent(Z, # matriz de linkage, en nuestro caso usamos ward
                      depth # niveles por debajo
                     )
```

```python
# [promedio, sd, count, factor de inconsistencia]
incons[-10:]
```

## Método del codo

```python
last = Z[-10:,2]
last_rev = last[::-1]
print(last_rev)
idx = np.arange(1, len(last)+1)
plt.plot(idx,last_rev)

acc=np.diff(last,2)
acc_rev=acc[::-1]
plt.plot(idx[:-2], acc_rev)
plt.show()

k = acc_rev.argmax()+2
print("El num óptimo de cluster es %s"%str(k))
```

## Parámetros distribución multivariada gaussian

$$ 
\mu = 
    \begin{bmatrix}
    \mu_x & \mu_y
    \end{bmatrix}
$$

$$ 
cov = 
    \begin{bmatrix}
    \sigma_{x,x} & \sigma_{x,y} \\
    \sigma_{y,x} & \sigma_{y,y}
    \end{bmatrix}
$$


```python
mean = [40,40]

cov = [
        [20,1], # cov x
        [1,30] # cov y
      ]
c = np.random.multivariate_normal(mean, cov, size=[200,])
d = np.random.multivariate_normal([80,80], [[30,1],[1,30]], size=[200,])
e = np.random.multivariate_normal([0,100], [[100,1],[1,100]], size=[200,])
X2 = np.concatenate((X,c,d,e))
plt.scatter(X2[:,0],X2[:,1])
plt.show()
```

```python
Z = linkage(X2, "ward")
```

```python
plt.figure(figsize=(10,10))
dendogram_tune(
    Z,
    truncate_mode="lastp",
    p=30,
    leaf_rotation=10.,
    leaf_font_size=10.,
    show_contracted=True,
    annotate_above=40,
    max_d=170
)
plt.show()
```

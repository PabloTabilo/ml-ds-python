# SVM Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.sort(5*np.random.rand(200,1),axis=0)
Y = np.sin(X).ravel()
Y[::5] += 3*(.5 - np.random.rand(40))
```

```python
plt.scatter(X,Y,color="darkorange",label="data")
```

```python
from sklearn.svm import SVR
```

```python
C=1e3
svr_lin=SVR(kernel="linear",C=C)
svr_rbg=SVR(kernel="rbf",C=C,gamma=.1)
svr_poly=SVR(kernel="poly",C=C,degree=3)
```

```python
y_lin = svr_lin.fit(X,Y).predict(X)
y_rbf = svr_rbg.fit(X,Y).predict(X)
y_poly = svr_poly.fit(X,Y).predict(X)
```

```python
lw = 2
plt.figure(figsize=(16,9))
plt.scatter(X,Y,color="darkorange",label="data")
plt.plot(X,y_lin,color="navy",lw=lw,label="SVM Linear")
plt.plot(X,y_rbf,color="c",lw=lw,label="SVM RBF")
plt.plot(X,y_poly,color="cornflowerblue",lw=lw,label="SVM Poly")
plt.xlabel("x")
plt.ylabel("y")
plt.title("SVR")
plt.legend()
plt.show()
```
